```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load hanya kolom penting
df = pd.read_parquet("yellow_tripdata_2025-01.parquet", columns=["trip_distance", "fare_amount"])

# Sampling
sample = df.sample(500, random_state=42)

# Histogram
sns.histplot(sample["trip_distance"], kde=False)
plt.show()

```

```python
# Visualisasi 2: Boxplot untuk trip_distance
plt.figure(figsize=(10, 6))
sns.boxplot(x=df['trip_distance'])
plt.title('Boxplot Trip Distance')
plt.xlabel('Trip Distance (miles)')
plt.show()
```

```python
# Cek kolom yang ada
print(sample.columns)

# Misal nama kolom yang benar adalah 'fare_amount'
plt.figure(figsize=(8,5))
sns.kdeplot(data=sample, x="fare_amount", fill=True)
plt.title("Density Plot Fare Amount")
plt.xlabel("Fare Amount ($)")
plt.show()
```
